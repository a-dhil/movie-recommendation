```python
# Import Dependencies
import pandas as pd
from pathlib import Path
import re
import matplotlib.pyplot as plt
```

```python
# Create a path to csv and read it in pandas dataframe
path=Path("Resources/movies_metadata.csv")
movies_df=pd.read_csv(path)
movies_df.head()
```

```python
movies_df.info()
```

```python
reduced_movies_df = movies_df.loc[:, ["id","title","genres"]]
reduced_movies_df
```

```python

reduced_movies_df = reduced_movies_df.rename(columns={"id": "movieId"
                                                })
reduced_movies_df.head()
```

```python
# Define a function to clean the genre column
def clean_genre(genre_string):
    # Find all instances of genre names using regular expression
    genres = re.findall(r"'name':\s*'([^']*)'", genre_string)
    # Join the genres with a '|' separator
    return '|'.join(genres)
```

```python
reduced_movies_df = reduced_movies_df.copy()

# Apply the function to the genre column
reduced_movies_df['cleaned_genres'] = reduced_movies_df['genres'].apply(clean_genre)

# Display the DataFrame to verify the results
# If you want to display both original and cleaned genres
print(reduced_movies_df[['cleaned_genres']])
```

```python
# Create a path to csv and read it in pandas dataframe
path=Path("Resources/ratings_small.csv")
movies_df=pd.read_csv(path)
movies_df.head()
```

```python
#movies_df.info()
movies_df["rating"].value_counts()
```

```python
n_ratings = len(movies_df)
n_movies = len(movies_df['movieId'].unique())
n_users = len(movies_df['userId'].unique())
 
print(f"Number of ratings: {n_ratings}")
print(f"Number of unique movieId's: {n_movies}")
print(f"Number of unique users: {n_users}")
print(f"Average ratings per user: {round(n_ratings/n_users, 2)}")
print(f"Average ratings per movie: {round(n_ratings/n_movies, 2)}")
```

```python
non_integer_rows = reduced_movies_df[pd.to_numeric(reduced_movies_df['movieId'], errors='coerce').isna()]
```

```python
reduced_movies_df = reduced_movies_df[pd.to_numeric(reduced_movies_df['movieId'], errors='coerce').notna()]
```

```python
reduced_movies_df['movieId'] = reduced_movies_df['movieId'].astype(int)
```

```python
movie_rating_df = pd.merge(reduced_movies_df, movies_df, on=["movieId"])
movie_rating_df.head()
```

```python
# Now, drop the 'genres' column from the DataFrame
movie_rating_df = movie_rating_df.drop(columns=['genres'])
movie_rating_df
```

```python
n_ratings = len(movie_rating_df)
n_movies = len(movie_rating_df['movieId'].unique())
n_users = len(movie_rating_df['userId'].unique())
 
print(f"Number of ratings: {n_ratings}")
print(f"Number of unique movieId's: {n_movies}")
print(f"Number of unique users: {n_users}")
print(f"Average ratings per user: {round(n_ratings/n_users, 2)}")
print(f"Average ratings per movie: {round(n_ratings/n_movies, 2)}")
```

```python
user_freq = movie_rating_df[['userId', 'movieId']].groupby(
    'userId').count().reset_index()
user_freq.columns = ['userId', 'n_ratings']
print(user_freq.head())
```

```python
movie_freq = movie_rating_df[['movieId', 'rating']].groupby('movieId').count().reset_index()
movie_freq.columns = ['movieId', 'n_ratings']
print(movie_freq)
```
